```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# First, we get some data

We generate some random data just to demonstrate how Matplotlib works.

We'll work with real data later to show how Matplotlib can help use explore the signals and structures in the data.

```python
np.random.seed(19680801)
n_obs = 200
RW = pd.DataFrame(
    dict(
        w1 = np.cumsum(np.random.randn(n_obs)),
        w2 = np.cumsum(np.random.randn(n_obs))
    )
)
RW.index.name = 't'
```

```python
RW
```

# Method 1: Pure Matplotlib

First, we create Figure and Axes objects by calling the `subplots` function.

- A **Figure** object is the container that holds the plots. It can have one or more plots in it.

- An **Axes** object is the container for an individual plot. 

Next, we map our data to the axes objects. 

We are going to create two objects, one for each columns of data.

Finally, we add some language and stuff.

```python
# Create the figure and axes objects
# Note that we don't use the fig object in this example
fig, ax = plt.subplots(figsize=(6, 4))

# Map data onto the axes in a particular way
ax.plot(RW.index, RW.w1, label='Random walk 1')
ax.plot(RW.index, RW.w2, label='Random walk 2')

# Add some labeling
ax.set_xlabel('Time [s]')
ax.set_ylabel('Distance [km]')
ax.set_title('Random walk example')
ax.legend();
```

# Method 2: In Combination with Pandas

We use Panda's built-in plot methods with `plt`.

Note that here we interact directly with the `plt` object.

Pandas handles our `Axes` object. 

```python
# First, we define some attributes for our figure
plt.figure(figsize=(6, 4));

# Next, we use Pandas to plot the data onto our figure and axes
RW.w1.plot(label="Random Walk 1", legend=True)
RW.w2.plot(label="Random Walk 2", legend=True);

# Then we add some labels
plt.xlabel("Time [s]")
plt.ylabel("Distance [km]")
plt.title("Random Walk Example")
```

# Method 3: Pandas Alone

We use Panda's built-in plot methods. 

Note that the arguments in the second plot apply to the figure as a whole.

```python
RW.w1.plot(figsize=(6,4), legend=True, label="Random Walk 1")
RW.w2.plot(xlabel="Time [s]", ylabel="Distance [km]", title="Random Walk Example", legend=True, label="Random Walk 2");
```
