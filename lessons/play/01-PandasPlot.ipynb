# Iris

```python
iris = pd.read_csv("iris.csv")
```

```python
iris.plot.line();
```

What does the $x$ axis represent?

```python
my_x = 'sepal.length'
my_y = 'sepal.width'
```

```python
ax = iris.plot(my_x, my_y)
```

```python
my_x = 'sepal.length'
my_y = 'sepal.width'
ax = iris.plot(my_x, my_y, style='o')
```

```python
ax = iris.plot.scatter(my_x, my_y)
```

What are the differences?

Also, what does the $x$ axis represent now?

```python
my_x = 'sepal.length'
my_y = 'sepal.width'
colors = {'Setosa':'red', 'Versicolor':'blue', 'Virginica':'green'}
ax = iris.plot.scatter(my_x, my_y, color=[colors[v] for v in iris.variety.to_list()]);
```

```python
iris.plot.line(style='.');
```
